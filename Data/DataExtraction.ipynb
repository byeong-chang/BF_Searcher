# 데이터 호출 및 필요 없는 Column 제거 및 몇몇 Column 이름 변경

```python
import pandas as pd
df = pd.read_csv('BarrierFree.csv') 
df.loc[df['시도 명칭'] == '세종특별자치시', '시군구 명칭'] = '세종특별자치시' # 세종 특별 자치시 시군구 명칭 null이라서 채워줌
col = ['시설명', '카테고리2', '시군구 명칭', '위도', '경도', '지번주소',
    '전화번호', '홈페이지', '휴무일', '운영시간',
    '무료주차 가능여부', '유료주차 가능여부', '장애인용 출입문', '휠체어 대여 가능 여부',
    '장애인 화장실 유무', '장애인 전용 주차장 여부', '대형주차장 가능여부', '시각장애인 안내견 동반 가능 여부',
    '점자 가이드 여부', '오디오 가이드(한국어)']
df = df[col]
df.rename(columns={'카테고리2':'분류'},inplace = True)
df
```

# 카테고리 데이터 있음 =1, 없음 = 0 으로 치환

```python

df['무료주차 가능여부'] = df['무료주차 가능여부'].replace('무료주차 불가',0)
df['무료주차 가능여부'] = df['무료주차 가능여부'].replace('무료주차 가능',1)

df['유료주차 가능여부'] = df['유료주차 가능여부'].replace('유료주차 불가',0)
df['유료주차 가능여부'] = df['유료주차 가능여부'].replace('유료주차 가능',1)

df['휠체어 대여 가능 여부']=df['휠체어 대여 가능 여부'].replace('휠체어 대여 불가',0)
df['휠체어 대여 가능 여부']=df['휠체어 대여 가능 여부'].replace('휠체어 대여 가능',1)

df['장애인용 출입문']=df['장애인용 출입문'].replace('장애인용 출입문 있음',1)
df['장애인용 출입문']=df['장애인용 출입문'].replace('장애인용 출입문 없음',0)

df['장애인 화장실 유무']=df['장애인 화장실 유무'].replace('장애인 화장실 있음',1)
df['장애인 화장실 유무']=df['장애인 화장실 유무'].replace('장애인 화장실 없음',0)

df['장애인 전용 주차장 여부']=df['장애인 전용 주차장 여부'].replace('장애인 주차구역 있음',1)
df['장애인 전용 주차장 여부']=df['장애인 전용 주차장 여부'].replace('장애인 주차구역 없음',0)

df['대형주차장 가능여부']=df['대형주차장 가능여부'].replace('대형차 주차 가능',1)
df['대형주차장 가능여부']=df['대형주차장 가능여부'].replace('대형차 주차 불가',0)

df['시각장애인 안내견 동반 가능 여부']=df['시각장애인 안내견 동반 가능 여부'].replace('시각장애인 안내견 동반 가능',1)
df['시각장애인 안내견 동반 가능 여부']=df['시각장애인 안내견 동반 가능 여부'].replace('시각장애인 안내견 동반 불가',0)

df['점자 가이드 여부']=df['점자 가이드 여부'].replace('장애인 안내 점자 있음',1)
df['점자 가이드 여부']=df['점자 가이드 여부'].replace('장애인 안내 점자 없음',0)

df['오디오 가이드(한국어)']=df['오디오 가이드(한국어)'].replace('오디오 가이드 있음(한국어)',1)
df['오디오 가이드(한국어)']=df['오디오 가이드(한국어)'].replace('오디오 가이드 없음(한국어)',0)
df
```

```python
# 분류 정규화
CultureCategory = df['분류'].unique()
for i,val in enumerate(CultureCategory):
    df['분류'].replace(val,i+1,inplace=True)
print(CultureCategory)
```

```python
# 시군구 명칭 정규화
LocationCategory = df['시군구 명칭'].unique()
for i,val in enumerate(LocationCategory):
    df['시군구 명칭'].replace(val,i+1,inplace=True)
print(LocationCategory)
df
```
